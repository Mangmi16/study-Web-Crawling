### 검색어 + time stamp로 폴더 생성해주는 함수

```python
def create_folder(query_txt):
    #현재 시간 저장하기
    now = time.localtime() #현재시간 now로 저장하기
    s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

    f_dir = 'c:\\temp\\' #입력받은 경로

    print(f_dir+query_txt+'-'+s)

    import os

    os.makedirs(f_dir+query_txt+'-'+s) 
    os.chdir(f_dir+query_txt+'-'+s)
    # 크롤링 결과 저장할 폴더 이름
    f_dir_final = f_dir+query_txt+'-'+s
    return f_dir_final 
```

### 이미지 추출함수 만들어 사이트 크롤러 적용하기

```python
def img_downloader(soup):
    file_no = 0 
    # 원하는 사진들이 있는 body 지정
#     html = driver.page_source
#     soup = BeautifulSoup(html, 'html.parser')
    title = soup.find('div', 'tit').text.strip()
    print(title, '이미지 추출중...')
    img_src = soup.find("div", 'box_txtPhoto')
    # 지정한 바디에서 'img' 태그값을 가지는 요소들 모두 데려온다
    img_src = img_src.find_all("img")

    # 추출한 이미지html 태그를 img_src_list에 담아 준다
    for i in img_src :
        # Src 태그에 해당되는 태그값만 img_src_list에 담아 준다
        img_src_list.append(i['src']) 
#         count += 1
        
        # 각 링크에서 이미지 다운받아오기. 
    for i in range(0,len(img_src_list)) :
        try :
            urllib.request.urlretrieve(img_src_list[i], title + '_' + str(file_no)  + '.jpg')

        except :
            print('추출실패')
            continue   
            
        file_no += 1                
        time.sleep(0.5)  
        print("%s번째 이미지 저장중.." %file_no)
```

### 상세 페이지 크롤링 함수

```python
# 문제4 : 상세 페이지 정보 긁어오기
def single_page (no):
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    head = soup.find('div', class_="tit_cont titleType1")
    body = soup.find('div', class_="box_txtPhoto")
    
    


    try:
        title = head.find('h2').text.strip()
        print("title 추출 완료")
        img_downloader(soup) ##### 이미지 다운로더 함수 삽입!
    except:
        print("내용이 없습니다")
    else:
        region_date = head.find('div','area_address').find_all('span')
        region = region_date[0].text.strip()
        date = region_date[1].text.strip()
        likes = head.find('span',id="conLike").text.strip()
        views = head.find('span', id="conRead").text.strip()
        body = soup.find('div', class_="box_txtPhoto").text.strip()


        print("article no : ", no)
        print("제목 : ", title)
        print("지역 : ", region)
        print("날짜 : ", date)
        print("좋아요 : ", likes)
        print("조회수 : ", views)
        print("본문 : ", body)
        print("=-"*60)

        title_li_idv.append(title)
        contents_li_idv.append(body)
        date_li_idv.append(date)
        likes_li_idv.append(likes)
        views_li_idv.append(views)
        #지역
        region_li_idv.append(region)
        #해쉬태그
        no += 1
        
    return no 
```

```python
# 문제 5 : 상세 페이지 하나씩 들어갔다 나오기
for title_no in range(1,12):
    time.sleep(1)
    try:
#     driver.find_element_by_xpath("""//*[@id="listBody"]/ul/li[%s]/div[2]/div[1]/a""" % title_no).click()
        driver.find_element_by_css_selector('#listBody > ul > li:nth-child(%s) > div.area_txt > div.tit > a' %title_no).click( )
        time.sleep(2)
        print("article no:", no)
        
        #상세 페이지 크롤링 적용
        no = single_page(no)
        if no == cnt:
            break
        print("page : " , title_no)
        
    except:
        continue
    driver.back()
```
